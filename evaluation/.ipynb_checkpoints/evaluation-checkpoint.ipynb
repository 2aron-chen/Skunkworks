```python
T1Dict = {
    "Whole Brain" : [wb_2min_dl, wb_9min_dl, wb_gt],
    "Gray Matter" : [gm_2min_dl, gm_9min_dl, gm_gt],
    "White Matter" : [wm_2min_dl, wm_9min_dl, wm_gt],
}
```

```python
fig, ax = plt.subplots(2, 3, figsize=(20, 16))
colors = sns.color_palette("rocket")

for i, name in enumerate(T1Dict.keys()):
    
    pred1, pred2, true = T1Dict[name]
    
    percent_dif = 100*(np.array(pred1)-np.array(true))/np.array(true)
    ax[0,i].axhline(0, linestyle="--", color='black')
    ax[0,i].set_title(f'[Denoised DL] {name} Percent Difference\nPvalue = {round(scipy.stats.f_oneway(pred1, true)[1], 4)}')
    sns.scatterplot(ax=ax[0,i], x=true, y=percent_dif, color=colors[i])
    ax[0,i].set_xlabel('T1 in second') 
    ax[0,i].set_ylabel('Percent difference')
    #ax[0,i].set_ylim([-15, 15])
    m, c, r, p, std_err = scipy.stats.linregress(true, percent_dif)
    x = np.linspace(np.min(true), np.max(true), 100)
    y = m*x+c
    ax[0,i].plot(x, y, color = 'red', label=f'R2={round(r, 2)}')
    ax[0,i].legend()
    
    
    percent_dif = 100*(np.array(pred2)-np.array(true))/np.array(true)
    ax[1,i].axhline(0, linestyle="--", color='black')
    ax[1,i].set_title(f'[Full DL] {name} Percent Difference\nPvalue = {round(scipy.stats.f_oneway(pred2, true)[1], 4)}')
    sns.scatterplot(ax=ax[1,i], x=true, y=percent_dif, color=colors[i])
    ax[1,i].set_xlabel('T1 in second') 
    ax[1,i].set_ylabel('Percent difference')
    #ax[1,i].set_ylim([-15, 15])
    m, c, r, p, std_err = scipy.stats.linregress(true, percent_dif)
    x = np.linspace(np.min(true), np.max(true), 100)
    y = m*x+c
    ax[1,i].plot(x, y, color = 'red', label=f'R2={round(r, 2)}')
    ax[1,i].legend()
    
plt.show()
```

```python
#smoothening
gpu_id = 'cuda'
smoother = GaussianSmoothing(1, 10, 0.2, 3).to(gpu_id)
def smoothen(pred, mask):
    e = np.copy(pred)
    e[np.expand_dims(mask, 0)==0] = 0.0
    e = smoother(torch.tensor(e).float().to(gpu_id))
    return e.cpu().numpy()
```
